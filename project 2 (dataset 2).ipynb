```python
def value_to_float(s):
    if 'c' in s:
        if len(s) > 1:
          return float(s.replace('c','')) * 95.0
    if 'd' in s:
        if len(s) > 1:
           return float(s.replace('d','')) * 80.0
    if 'b' in s:
        if len(s) > 1:
           return float(s.replace('b','')) * 70.0 
data['math score']=data['math score'].apply(value_to_float)
```
