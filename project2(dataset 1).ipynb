```python
def value_to_float(s):
    if '88' in s:
        if len(s) > 1:
          return float(s.replace('88','')) * 4.50
    if '200' in s:
        if len(s) > 1:
           return float(s.replace('200','')) * 5.850 
    if '500' in s:
        if len(s) > 1:
           return float(s.replace('500','')) * 8.80 
data['votes']=data['votes'].apply(value_to_float)
```
